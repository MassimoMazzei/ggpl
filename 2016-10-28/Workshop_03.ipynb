### Creating stairs:
This function create a u-shaped stairs inside a box of dx,dy and dz dimensions

```python
"""This function returns a hpc value containing a U-shaped stair"""
def ggpl_uShapedStair(dx,dy,dz):
    
    # Creating the first step
    numberOfSteps = int(dz/.18)
    riser = float(dz)/numberOfSteps
    tread = .64-2*riser
    step = CUBOID([float(dx)/2,tread,riser])
    understep = MKPOL([[[0,0,0],[float(dx)/2,0,0],[float(dx)/2,0,riser],[0,0,riser],[float(dx)/2,tread,riser],[0,tread,riser]],[[1,2,3,4,5,6],[]],1])
    step = STRUCT([step,T(2)(tread),understep])
    
    # Creating the first ramp and the landing
    ramp = STRUCT([step])
    for i in range(1,numberOfSteps/2):
        ramp = STRUCT([ramp, T(2)(tread), T(3)(riser), ramp])
    
    landing = CUBOID([dx,float(dy)/4,riser])
    stair = STRUCT([ramp,T(2)(tread*numberOfSteps/2),T(3)(riser*numberOfSteps/2),landing])
    ramp = STRUCT([R([1,2])(PI),T([1,2,3])([-float(dx),-tread*numberOfSteps/2,riser*numberOfSteps/2]),ramp])
    VIEW(STRUCT([ramp,stair]))
    step = CUBOID([float(dx)/2,tread,riser])
    
   

    
ggpl_uShapedStair(4,5,4)
```
